### (a)
$$ J(W) = \frac{\left | \widetilde{m_{1}}-\widetilde{m_{2}} \right |^{2}}{\widetilde{s}_{1}^{2} + \widetilde{s}_{2}^{2} }$$
<br>
In the above equation value of J changes with respect to W.That is if the value of W increases the value of J also increases and vice versa. 
<br>
$$ J(W) = \frac{W^{t}S_{B}W}{W^{t}S_{W}W}$$
<br>
Here, the value of W can change the values of above equation as well and It satisfies the equation where $S_{B}$ is between-class scatter matrix and  $S_{W}$ is within-class scatter matrix. Hence, it satisfies the below equation 
<br>
$$S_{B}W = \lambda S_{W}W$$



### (b)
Scatter matrix $S_{i}$ is defined for the multiple classes where $i$ is number of classes. 
<br>
<br>
$$ S_{i} =\sum_{x\epsilon D_{i}} (m_{1} - m_{2})(m_{1} - m_{2})^{t} $$
<br>
And the between - class for the same can be given as
<br>
<br>
$$ S_{B} = (m_{1} - m_{2})(m_{1} - m_{2})^{t} $$
<br>
Multiplyting the abov equation by $\frac{n1 * n2}{n}$  Gives us the following 

$$ S_{B} =\sum_{i=1}^{c} (m_{i} - m)(m_{i} - m)^{t} $$
